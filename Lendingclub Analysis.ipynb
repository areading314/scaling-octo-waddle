```python
from analyze import load_loan_data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df = load_loan_data()
```

```python
df
```

```python
df.describe()
```

```python
groups3 = df[df.term == 36].groupby('grade')
groups5 = df[df.term == 60].groupby('grade')
fig, axarr = plt.subplots(2, sharey=True)
for name, group in groups3:
    axarr[0].plot(100*group.int_rate, group.fico_range_low, marker='.', linestyle='', ms=12, label=name, alpha=0.8)

for name, group in groups5:
    axarr[1].plot(100*group.int_rate, group.fico_range_low, marker='.', linestyle='', ms=12, label=name, alpha=0.8)

for ax in axarr:
    ax.legend(numpoints=1, loc='upper right')

axarr[0].set_title('3 year term')
axarr[1].set_title('5 year term')
axarr[1].set_xlabel('Interest Rate')
plt.ylabel('Fico Score')
plt.show()
```
